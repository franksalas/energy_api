```python
def category(api_key=None, category_id=None):
    '''input: api_key, category_id
    return: dict of category'''
    
    url = 'http://api.eia.gov/category/'
    payload = {'api_key':api_key,
               'category_id':category_id}
    r = requests.get(url, params=payload)
    return r.json()
```
